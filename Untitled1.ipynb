```python
# ============================================================
# 📊 Hyundai Car Sales Data Visualization - Google Colab
# ============================================================

# --- Import libraries ---
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Enable inline plotting
%matplotlib inline

# --- Load Dataset ---
# If you have a CSV, upload it or replace the path with your own dataset.
# For demo, let's create a small synthetic dataset.

data = {
    "Model": ["Creta", "Venue", "i20", "Verna", "Aura", "Tucson", "Exter", "Alcazar"] * 12,
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"] * 8,
    "Year": [2024]*96,
    "Units_Sold": [4200, 3900, 3700, 2500, 1800, 900, 2100, 1600]*12,
    "Avg_Price": [1100000, 1000000, 950000, 1300000, 900000, 2800000, 800000, 2000000]*12
}

df = pd.DataFrame(data)

# --- Basic Data Analysis ---
print("First 5 rows of the dataset:")
display(df.head())

print("\nSummary Statistics:")
display(df.describe())

# ============================================================
# 🟩 Matplotlib Visualizations
# ============================================================

# 1. Bar Chart – Total Sales by Model
plt.figure(figsize=(10,6))
df.groupby("Model")["Units_Sold"].sum().sort_values(ascending=False).plot(kind='bar', color='skyblue')
plt.title("Total Hyundai Car Sales by Model")
plt.ylabel("Total Units Sold")
plt.xlabel("Model")
plt.grid(axis='y')
plt.show()

# 2. Histogram – Distribution of Average Prices
plt.figure(figsize=(8,5))
plt.hist(df["Avg_Price"], bins=10, color='lightgreen', edgecolor='black')
plt.title("Distribution of Average Car Prices")
plt.xlabel("Price (INR)")
plt.ylabel("Frequency")
plt.show()

# 3. Line Chart – Sales Trend (first 50 rows)
plt.figure(figsize=(10,5))
plt.plot(df["Units_Sold"].head(50), marker='o')
plt.title("Units Sold Trend (First 50 Records)")
plt.xlabel("Record Index")
plt.ylabel("Units Sold")
plt.grid(True)
plt.show()

# ============================================================
# 🟦 Seaborn Visualizations
# ============================================================

# 1. Boxplot – Price by Model
plt.figure(figsize=(10,6))
sns.boxplot(x="Model", y="Avg_Price", data=df, palette="cool")
plt.title("Boxplot of Average Price by Model")
plt.xticks(rotation=45)
plt.show()

# 2. Violin Plot – Units Sold by Model
plt.figure(figsize=(10,6))
sns.violinplot(x="Model", y="Units_Sold", data=df, palette="viridis")
plt.title("Violin Plot of Units Sold by Model")
plt.xticks(rotation=45)
plt.show()

# 3. Count Plot – Sales Count per Model
plt.figure(figsize=(10,5))
sns.countplot(x="Model", data=df, palette="Set2")
plt.title("Count of Records per Car Model")
plt.xticks(rotation=45)
plt.show()

# ============================================================
# 🟧 Interactive Visualizations with Plotly
# ============================================================

# 1. Interactive Scatter Plot – Units Sold vs Price
fig = px.scatter(df, x="Avg_Price", y="Units_Sold", color="Model",
                 size="Units_Sold", hover_name="Model",
                 title="Interactive Scatter Plot - Price vs Sales")
fig.show()

# 2. Interactive Histogram – Average Price Distribution
fig = px.histogram(df, x="Avg_Price", color="Model", nbins=10,
                   title="Interactive Histogram of Average Prices")
fig.show()

print("✅ Visualization Complete!")
```
